```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsforecast.models import CrostonOptimized

# Пример данных (интермиттирующий спрос)
data = pd.Series([0, 0, 5, 0, 0, 0, 10, 0, 0, 3, 0, 0, 0, 7, 0, 0, 2, 0, 0, 0])

# Преобразуем данные в numpy-массив
y = data.values

# Прогноз с использованием CrostonOptimized
croston_model = CrostonOptimized()
forecast_croston = croston_model.forecast(y, h=5)


# Визуализация
plt.figure(figsize=(12, 6))
plt.plot(range(len(y)), y, label="Исходные данные", marker='o', linestyle="--", color="black")
plt.axvline(x=len(y)-1, color='red', linestyle="--", label="Начало прогноза")

# Добавляем прогнозы
forecast_range = range(len(y), len(y) + 5)
plt.plot(forecast_range, forecast_croston['mean'], label="Croston", marker='o', color="blue")


plt.legend()
plt.title("Прогноз методом Croston и его модификаций")
plt.xlabel("Время")
plt.ylabel("Прогнозируемый спрос")
plt.show()
```
