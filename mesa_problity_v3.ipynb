```python
import pandas as pd
df_all = pd.read_excel('/content/mat_y_40_sku.xlsx')
df=df_all[(df_all['Month']==12) & (df_all['Analytic4']=='Ванночки') & (df_all['Year']==2023)].copy()
# df_12['Brand'] = df_12['Brand'].astype('category').cat.codes
df['weight'] = df['Analytic1'].str.replace(' г', '').astype(int)
df['share_vol_true'] =df['SALES VOL'].apply(lambda x: x/(df['SALES VOL'].sum()))
# df_12[['PRICE PER UNIT','Brand','Sku_name','weight','share_vol_true']]
# print(df_12.head(10).to_string())
print(df[['PRICE PER UNIT','Brand','Sku_name','weight','share_vol_true']].to_string())
```

```python
import torch
from torch.optim import Adam
from mesa import Agent, Model
from mesa.time import RandomActivation
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Фиксируем случайные семена для воспроизводимости
torch.manual_seed(0)
np.random.seed(0)

# Инициализируем параметры для оптимизации как скаляры
price_sensitivity = torch.tensor(0.5, requires_grad=True)
weight_sensitivity = torch.tensor(0.5, requires_grad=True)
brand_preference = torch.tensor(0.5, requires_grad=True)

# Оптимизатор
optimizer = Adam([price_sensitivity, weight_sensitivity, brand_preference], lr=0.1)

# # Данные из базы данных
# data = {
#     'PRICE PER UNIT': [469.424564, 449.267694, 816.116336, 720.726262, 580.349508, 581.037628, 483.201577, 670.511345, 734.084212, 480.710229, 665.564599, 577.558387, 638.083749, 720.935253, 306.674636, 710.796331, 691.699741, 358.868333],
#     'Brand': ['LiebenDorf', 'Свежий ряд', 'Фетакса', 'Hochland', 'Viola', 'President', 'Карат', 'Hochland', 'Viola', 'LiebenDorf', 'Hochland', 'President', 'Фетакса', 'Hochland', 'Ежедень', 'Hochland', 'Viola', 'Переяславль'],
#     'Sku_name': [
#         'Сыр плавленый LiebenDorf Сливочный Ванночки 400 г 55,0%',
#         'Сыр плавленый Свежий ряд Янтарь Ванночки 400 г 45,0%',
#         'Сыр плавленый Фетакса Ванночки 200 г 45,0%',
#         'Сыр плавленый Hochland Сливочный Ванночки 200 г 55,0%',
#         'Сыр плавленый Viola Сливочный Ванночки 400 г 50,0%',
#         'Сыр плавленый President Сливочный Ванночки 400 г 45,0%',
#         'Сыр плавленый Карат Дружба Ванночки 400 г 45,0%',
#         'Сыр плавленый Hochland Сливочный Ванночки 400 г 55,0%',
#         'Сыр плавленый Viola Сливочный Ванночки 200 г 50,0%',
#         'Сыр плавленый LiebenDorf с ветчиной Ванночки 400 г 55,0%',
#         'Сыр плавленый Hochland с ветчиной Ванночки 400 г 55,0%',
#         'Сыр плавленый President с ветчиной Ванночки 400 г 45,0%',
#         'Сыр плавленый Фетакса Ванночки 400 г 45,0%',
#         'Сыр плавленый Hochland с грибами Ванночки 400 г 55,0%',
#         'Сыр плавленый Ежедень Классический Ванночки 180 г 55,0%',
#         'Сыр плавленый Hochland с ветчиной Ванночки 200 г 55,0%',
#         'Сыр плавленый Viola с лисичками Ванночки 200 г 50,0%',
#         'Сыр плавленый Переяславль Сливочный Ванночки 400 г 25,0%'
#     ],
#     'weight': [400, 400, 200, 200, 400, 400, 400, 400, 200, 400, 400, 400, 400, 400, 180, 200, 200, 400],
#     'share_vol_true': [0.173770, 0.094276, 0.071347, 0.077681, 0.099023, 0.072950, 0.023941, 0.026313, 0.052944, 0.052605, 0.035156, 0.040442, 0.034191, 0.001701, 0.056206, 0.042792, 0.025751, 0.018909]
# }

# df = pd.DataFrame(data)

# Фактические исторические продажи для продуктов
historical_sales = torch.tensor(df['share_vol_true'].values, dtype=torch.float32)

# Нормализуем исторические продажи
max_sales = historical_sales.max()
min_sales = historical_sales.min()
normalized_historical_sales = (historical_sales - min_sales) / (max_sales - min_sales)

# Продукты с ценами, весом и брендами из DataFrame
products = []
unique_brands = df['Brand'].unique()
brand_to_index = {brand: index for index, brand in enumerate(df['Brand'].unique())}  # Создаем словарь для преобразования
num_brands = len(unique_brands)  # Общее количество уникальных брендов

for index, row in df.iterrows():

    products.append({
        "id": index,
        "brand": brand_to_index[row['Brand']] / (num_brands - 1),  # Заменяем строку на числовой индекс
        "cheese_type": row['Sku_name'],
        "price": row['PRICE PER UNIT'],
        "weight": row['weight']
    })

# Нормализуем признаки продуктов
max_price = max(product["price"] for product in products)
min_price = min(product["price"] for product in products)
max_weight = max(product["weight"] for product in products)
min_weight = min(product["weight"] for product in products)

for product in products:
    product["price_normalized"] = (product["price"] - min_price) / (max_price - min_price)
    product["weight_normalized"] = (product["weight"] - min_weight) / (max_weight - min_weight)

# Определяем класс агента
class CustomerAgent(Agent):
    def __init__(self, unique_id, model, parameters):
        super().__init__(unique_id, model)
        self.parameters = parameters  # Параметры чувствительности
        self.probabilities = None

    def step(self):
        # Получаем параметры
        price_sensitivity, weight_sensitivity, brand_preference = self.parameters

        # Вычисляем оценку для каждого продукта
        scores = []
        for product in self.model.products:
            price = torch.tensor(product["price_normalized"], dtype=torch.float32)
            weight = torch.tensor(product["weight_normalized"], dtype=torch.float32)
            brand = torch.tensor(product["brand"], dtype=torch.float32)  # Теперь это числовой индекс

            # Вычисляем оценку (скаляр)
            score = -price_sensitivity * price + weight_sensitivity * weight + brand_preference * brand
            scores.append(score)

        scores = torch.stack(scores)  # Форма [количество продуктов]

        # Применяем softmax для получения вероятностей
        probabilities = torch.softmax(scores, dim=0)  # Форма [количество продуктов]

        # Сохраняем вероятности для сбора данных
        self.probabilities = probabilities

# Определяем модель
class MarketModel(Model):
    def __init__(self, num_customers, products, parameters):
        self.num_customers = num_customers
        self.products = products
        self.parameters = parameters
        self.schedule = RandomActivation(self)
        self.datacollector = []

        for i in range(self.num_customers):
            agent = CustomerAgent(i, self, self.parameters)
            self.schedule.add(agent)

    def step(self):
        self.schedule.step()

        # Собираем вероятности от каждого агента
        total_probabilities = torch.zeros(len(self.products))
        for agent in self.schedule.agents:
            total_probabilities += agent.probabilities  # Теперь формы совпадают

        # Вычисляем средние вероятности
        average_probabilities = total_probabilities / self.num_customers
        self.datacollector.append(average_probabilities)

# Основной цикл оптимизации
num_epochs = 50
num_customers = 1000  # Увеличиваем число агентов для лучшей статистики

# Списки для хранения истории обучения
loss_history = []
price_sensitivity_history = []
weight_sensitivity_history = []
brand_preference_history = []
predicted_sales_history = []

for epoch in range(num_epochs):
    optimizer.zero_grad()

    # Создаём модель с текущими параметрами
    parameters = (price_sensitivity, weight_sensitivity, brand_preference)
    model = MarketModel(num_customers, products, parameters)

    # Запускаем модель на один шаг
    model.step()

    # Получаем средние вероятности покупки каждого продукта
    average_probabilities = model.datacollector[-1]

    # Предсказанные продажи (нормализованные)
    predicted_sales = average_probabilities

    # Вычисляем функцию потери
    loss = torch.mean((predicted_sales - normalized_historical_sales) ** 2)

    # Вычисляем градиенты
    loss.backward()

    # Обновляем параметры
    optimizer.step()

    # Ограничиваем параметры в диапазоне [0, 1]
    price_sensitivity.data.clamp_(0, 1)
    weight_sensitivity.data.clamp_(0, 1)
    brand_preference.data.clamp_(0, 1)

    # Сохраняем историю
    loss_history.append(loss.item())
    price_sensitivity_history.append(price_sensitivity.item())
    weight_sensitivity_history.append(weight_sensitivity.item())
    brand_preference_history.append(brand_preference.item())
    predicted_sales_history.append(predicted_sales.detach().numpy())

    # Выводим текущие значения
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss.item():.6f}, "
          f"Price Sensitivity: {price_sensitivity.item():.4f}, "
          f"Weight Sensitivity: {weight_sensitivity.item():.4f}, "
          f"Brand Preference: {brand_preference.item():.4f}")

# Построение графика функции потери
plt.figure(figsize=(10, 5))
plt.plot(loss_history)
plt.title('Изменение функции потери (Loss) по эпохам')
plt.xlabel('Эпоха')
plt.ylabel('Loss')
plt.grid(True)
plt.show()

# Построение графиков изменения параметров
plt.figure(figsize=(10, 5))
plt.plot(price_sensitivity_history, label='Price Sensitivity')
plt.plot(weight_sensitivity_history, label='Weight Sensitivity')
plt.plot(brand_preference_history, label='Brand Preference')
plt.title('Изменение параметров чувствительности по эпохам')
plt.xlabel('Эпоха')
plt.ylabel('Значение параметра')
plt.legend()
plt.grid(True)
plt.show()

# Сравнение предсказанных продаж с фактическими
# Берём предсказанные продажи из последней эпохи
predicted_sales_final = predicted_sales_history[-1]
# Де-нормализуем предсказанные продажи
denormalized_predicted_sales = predicted_sales_final * (max_sales.item() - min_sales.item()) + min_sales.item()

actual_sales = historical_sales.numpy()

# Вычисление метрик R2 и WAPE
def r2_score(actual, predicted):
    residual_sum_of_squares = np.sum((actual - predicted) ** 2)
    total_sum_of_squares = np.sum((actual - np.mean(actual)) ** 2)
    return 1 - (residual_sum_of_squares / total_sum_of_squares)

def wape(actual, predicted):
    return np.sum(np.abs(actual - predicted)) / np.sum(actual)

r2 = r2_score(actual_sales, denormalized_predicted_sales)
wape_value = wape(actual_sales, denormalized_predicted_sales)

# Визуализация результатов
plt.figure(figsize=(10, 5))
index = np.arange(len(actual_sales))
bar_width = 0.35

plt.bar(index, actual_sales, bar_width, label='Фактический доля рынка')
plt.bar(index + bar_width, denormalized_predicted_sales, bar_width, label='Предсказанные доля рынка')

plt.xlabel('Продукт')
plt.ylabel('Доля рынка')
plt.title('Сравнение фактических и предсказанных доля рынка')
plt.xticks(index + bar_width / 2, [p["cheese_type"] for p in products],rotation=90)
plt.legend()
plt.grid(True)
plt.show()

# Выводим фактические и предсказанные продажи и метрики
print("Фактические и предсказанные продажи:")
for i, product in enumerate(products):
    print(f"Продукт: {product['cheese_type']}")
    print(f"Фактические продажи: {actual_sales[i]:.2f}")
    print(f"Предсказанные продажи: {denormalized_predicted_sales[i]:.2f}\n")

# Выводим метрики
print(f"R2 Score: {r2:.4f}")
print(f"WAPE: {wape_value:.4f}")
```
