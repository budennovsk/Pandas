```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize

pd.options.display.float_format = '{:_.2f}'.format


# Данные
data = {
    'Цена': [400, 400, 400, 400, 400, 400, 400, 400, 400, 400, 400, 400],
    'СС': [179.77, 185.22, 183.18, 184.64, 185.25, 184.34, 179.62, 186.65, 190.82, 197.13, 209.59, 218.64],
    'Дистр': [0.50, 0.50, 0.50, 0.50, 0.50, 0.50, 0.50, 0.50, 0.50, 0.50, 0.50, 0.50]
}

# Создание DataFrame
df = pd.DataFrame(data)

# Определение целевой функции
def objective(x):

    # x содержит значения для x0 и x1
    x0 = x[:len(df)]
    x1 = x[len(df):]


    return -np.sum((np.exp(x0*(-0.00733681)) * 1094629.19 * x1 )* ((x0*0.65) - df['СС']))

# Определение ограничения
def constraint(x):
    # x содержит значения для x0 и x1
    x0 = x[:len(df)]
    x1 = x[len(df):]
    return 500000 - np.sum(np.exp(x0*(-0.00733681)) * 1094629.19 * x1)

# Начальные значения (в исходном DataFrame)
x0_init = df['Цена'].values
x1_init = df['Дистр'].values
x_init = np.concatenate([x0_init, x1_init])

# Границы для переменных (x0 - от 400 до 600, x1 - от 0 до 1)
bounds_x0 = [(400, 6000)] * len(df)
bounds_x1 = [(0, 1)] * len(df)
bounds = bounds_x0 + bounds_x1

# Ограничения
constraints = [{'type': 'ineq', 'fun': constraint}]

# Решение задачи оптимизации
solution = minimize(objective, x_init, bounds=bounds, constraints=constraints,  method='trust-constr') #constraints=constraints,  method='trust-constr' method='SLSQP'

# Оптимизированные значения
optimized_x0 = solution.x[:len(df)]
optimized_x1 = solution.x[len(df):]

# Вывод результатов
df['Цена_оптим'] = optimized_x0
df['Дистр_оптим'] = optimized_x1
df['Обьем_оптим'] =np.exp((-0.00733681)*df['Цена_оптим'])*df['Дистр_оптим']*1094629.19
df['Маржа_оптим']=df['Обьем_оптим']*((df['Цена_оптим']*0.65)-df['СС'])

df['Обьем_оптим_сумм'] = df['Обьем_оптим'].sum()
df['Маржа_оптим_сумм']=df['Маржа_оптим'].sum()


df

```
