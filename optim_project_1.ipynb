```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize

pd.options.display.float_format = '{:_.2f}'.format


# Данные для DataFrame
data = {
    'Цена': [400, 400, 400, 400, 400, 400, 400, 400, 400, 400, 400, 400],
    'СС': [179.77, 185.22, 183.18, 184.64, 185.25, 184.34, 179.62, 186.65, 190.82, 197.13, 209.59, 218.64],
}

# Создание DataFrame
df = pd.DataFrame(data)

# Функция для оптимизации (минимизация отрицательной маржи)
def objective(x):
    df['Цена_после_оптимизации'] = x
    df['Маржа_после_оптимизации'] = (np.exp(-df['Цена_после_оптимизации']*0.008)*931604) * ((df['Цена_после_оптимизации']*0.65) - df['СС'])

    return -df['Маржа_после_оптимизации'].sum()

# Начальные значения для цен
x0 = df['Цена'].values

# Ограничения
bounds = [(400, 600)] * len(x0)

# Оптимизация
result = minimize(objective, x0, method ='Nelder-Mead',bounds=bounds)

# Обновление цен в DataFrame
df['Цена_после_оптимизации'] = result.x
# df['Объем']= result.


# Пересчет маржи
# df['Маржа_после_оптимизации'] = (np.exp(-df['Цена_после_оптимизации']*0.008)*931604) * ((df['Цена_после_оптимизации']*0.65) - df['СС'])

df['Объем_после_оптимизации'] =  (np.exp(-df['Цена_после_оптимизации']*0.008)*931604)
df['Сумма_Маржа'] = df['Маржа_после_оптимизации'].sum()

# Вывод результата
df
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
from scipy.optimize import differential_evolution
import time

# Данные для DataFrame
data = {
    'Цена': [400, 400, 400, 400, 400, 400, 400, 400, 400, 400, 400, 400],
    'СС': [179.77, 185.22, 183.18, 184.64, 185.25, 184.34, 179.62, 186.65, 190.82, 197.13, 209.59, 218.64],
}

# Создание DataFrame
df = pd.DataFrame(data)

# Функция для оптимизации (минимизация отрицательной маржи)
def objective(x):
    df['Цена_после_оптимизации'] = x
    df['Маржа_после_оптимизации'] = (np.exp(-df['Цена_после_оптимизации']*0.008)*931604) * ((df['Цена_после_оптимизации']*0.65) - df['СС'])
    return -df['Маржа_после_оптимизации'].sum()

# Начальные значения для цен
x0 = df['Цена'].values

# Ограничения
bounds = [(400, 600)] * len(x0)

# Методы оптимизации
methods = ['Nelder-Mead', 'Powell', 'CG', 'BFGS', 'L-BFGS-B', 'SLSQP','differential_evolution']

# Результаты
results = []

for method in methods:
    if method == 'CG':
        start_time = time.time()
        result = minimize(objective, x0, method=method)
        sm =  np.sum((np.exp(-df['Цена_после_оптимизации']*0.008)*931604) * ((df['Цена_после_оптимизации']*0.65) - df['СС']))
        end_time = time.time()
        elapsed_time = end_time - start_time
        results.append({'метод': method, 'время': elapsed_time, 'маржа_сумма': sm,'результат': result.x})
    elif method == 'BFGS':
        start_time = time.time()
        result = minimize(objective, x0, method=method)
        sm =  np.sum((np.exp(-df['Цена_после_оптимизации']*0.008)*931604) * ((df['Цена_после_оптимизации']*0.65) - df['СС']))
        end_time = time.time()
        elapsed_time = end_time - start_time
        results.append({'метод': method, 'время': elapsed_time,'маржа_сумма': sm, 'результат': result.x})
    elif method == 'differential_evolution':
        start_time = time.time()
        result = differential_evolution(objective, bounds)
        sm =  np.sum((np.exp(-df['Цена_после_оптимизации']*0.008)*931604) * ((df['Цена_после_оптимизации']*0.65) - df['СС']))
        end_time = time.time()
        elapsed_time = end_time - start_time
        results.append({'метод': method, 'время': elapsed_time, 'маржа_сумма': sm,'результат': result.x})

    else:

        start_time = time.time()
        result = minimize(objective, x0, method=method, bounds=bounds)
        sm =  np.sum((np.exp(-df['Цена_после_оптимизации']*0.008)*931604) * ((df['Цена_после_оптимизации']*0.65) - df['СС']))
        end_time = time.time()
        elapsed_time = end_time - start_time
        results.append({'метод': method, 'время': elapsed_time,'маржа_сумма': sm, 'результат': result.x})

# Создание DataFrame с результатами
res = pd.DataFrame(results)

# Вывод результата
res
```
