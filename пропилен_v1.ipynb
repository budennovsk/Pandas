```python
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import GroupKFold
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.inspection import permutation_importance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Подготовка данных
# X = df.drop(['Plotnost__avg', 'ts_group','ProchnostNaSzhatie__avg','alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg','ptr_fluidity_low'], axis=1)
# X = df[['System1:RZN.L3.EX.ZE.TZZE.Data.ZEActValTempZone8_2__avg',
#        'System1:RZN.L3.EX.ZE.TZZE.Data.ZEActValTempZone8_3__avg',
#        'System1:RZN.L3.EX.ZE.TZZE.Data.ZEActValTempZone8_4__avg',
#         'Nomenklatura']]
X = df[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg','Nomenklatura']]
y = df['Plotnost__avg']

# Группы для кросс-валидации
groups = df['pres'].values

# # Преобразуем 'Nomenklatura' в категорический тип
# X['Nomenklatura'] = X['Nomenklatura'].astype('category')

# # Получаем индексы категориальных столбцов
# categorical_features = [X.columns.get_loc('Nomenklatura')]

print(f"Признаки: {X.shape[1]}")
# print(f"Категориальные признаки: {categorical_features}")
print(f"Уникальные группы: {len(np.unique(groups))}")

# GroupKFold для кросс-валидации по группам
gkf = GroupKFold(n_splits=5)

scores = {
    'train_rmse': [], 'val_rmse': [], 'test_rmse': [],
    'train_r2': [], 'val_r2': [], 'test_r2': [],
    'train_mae': [], 'val_mae': [], 'test_mae': [],
    'best_iteration': []
}

# Для хранения Permutation Importance
all_permutation_importance = []

# Кросс-валидация
for fold, (train_val_idx, test_idx) in enumerate(gkf.split(X, y, groups)):
    print(f"\n{'='*70}")
    print(f"--- Fold {fold + 1} ---")
    print('='*70)

    # Разделяем на train_val и test
    X_train_val, X_test = X.iloc[train_val_idx], X.iloc[test_idx]
    y_train_val, y_test = y.iloc[train_val_idx], y.iloc[test_idx]
    groups_train_val = groups[train_val_idx]

    # Внутренний split для train и val
    gkf_inner = GroupKFold(n_splits=4)
    for train_idx, val_idx in gkf_inner.split(X_train_val, y_train_val, groups_train_val):
        X_train, X_val = X_train_val.iloc[train_idx], X_train_val.iloc[val_idx]
        y_train, y_val = y_train_val.iloc[train_idx], y_train_val.iloc[val_idx]
        break  # Берем только первый split

    # Создаем Pool для CatBoost
    # train_pool = Pool(X_train, y_train, cat_features=categorical_features)
    # val_pool = Pool(X_val, y_val, cat_features=categorical_features)
    # test_pool = Pool(X_test, y_test, cat_features=categorical_features)
    train_pool = Pool(X_train[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']], y_train)
    val_pool = Pool(X_val[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']], y_val )
    test_pool = Pool(X_test[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']], y_test)

    # Обучаем модель с early stopping
    model = CatBoostRegressor(
        iterations=1000,  # Максимальное количество итераций
        learning_rate=0.1,
        depth=6,
        loss_function='RMSE',
        eval_metric='RMSE',
        early_stopping_rounds=50,  # Останавливаем если нет улучшений за 50 итераций
        random_state=42,
        verbose=False
    )

    model.fit(train_pool, eval_set=val_pool)

    # Получаем количество использованных итераций
    best_iteration = model.get_best_iteration()
    scores['best_iteration'].append(best_iteration)

    # Предсказания
    y_train_pred = model.predict(X_train[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']])
    y_val_pred = model.predict(X_val[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']])
    y_test_pred = model.predict(X_test[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']])

    # Метрики
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    train_r2 = r2_score(y_train, y_train_pred)
    val_r2 = r2_score(y_val, y_val_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    train_mae = mean_absolute_error(y_train, y_train_pred)
    val_mae = mean_absolute_error(y_val, y_val_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)

    scores['train_rmse'].append(train_rmse)
    scores['val_rmse'].append(val_rmse)
    scores['test_rmse'].append(test_rmse)
    scores['train_r2'].append(train_r2)
    scores['val_r2'].append(val_r2)
    scores['test_r2'].append(test_r2)
    scores['train_mae'].append(train_mae)
    scores['val_mae'].append(val_mae)
    scores['test_mae'].append(test_mae)

    print(f"Best iteration: {best_iteration}")
    print(f"Train RMSE: {train_rmse:.4f} | Val RMSE: {val_rmse:.4f} | Test RMSE: {test_rmse:.4f}")
    print(f"Train R²:   {train_r2:.4f} | Val R²:   {val_r2:.4f} | Test R²:   {test_r2:.4f}")
    print(f"Train MAE:  {train_mae:.4f} | Val MAE:  {val_mae:.4f} | Test MAE:  {test_mae:.4f}")

    # Permutation Importance на тестовом наборе
    print(f"\nРассчитываю Permutation Importance...")
    perm_importance = permutation_importance(
        model, X_test[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']], y_test,
        n_repeats=10,
        random_state=42,
        n_jobs=-1
    )

    perm_importance_df = pd.DataFrame({
        'feature': X[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg']].columns,
        'importance_mean': perm_importance.importances_mean,
        'importance_std': perm_importance.importances_std
    }).sort_values('importance_mean', ascending=False)

    all_permutation_importance.append(perm_importance_df)

    print(f"\nТоп-10 важных признаков (Permutation Importance) для Fold {fold + 1}:")
    print(perm_importance_df.head(10).to_string(index=False))

# Итоговые результаты
print("\n" + "="*70)
print("ИТОГОВЫЕ РЕЗУЛЬТАТЫ КРОСС-ВАЛИДАЦИИ")
print("="*70)
print(f"Лучшая итерация: {np.mean(scores['best_iteration']):.1f} ± {np.std(scores['best_iteration']):.1f}")
print()
print(f"RMSE (Train): {np.mean(scores['train_rmse']):.4f} ± {np.std(scores['train_rmse']):.4f}")
print(f"RMSE (Val):   {np.mean(scores['val_rmse']):.4f} ± {np.std(scores['val_rmse']):.4f}")
print(f"RMSE (Test):  {np.mean(scores['test_rmse']):.4f} ± {np.std(scores['test_rmse']):.4f}")
print()
print(f"R² (Train):   {np.mean(scores['train_r2']):.4f} ± {np.std(scores['train_r2']):.4f}")
print(f"R² (Val):     {np.mean(scores['val_r2']):.4f} ± {np.std(scores['val_r2']):.4f}")
print(f"R² (Test):    {np.mean(scores['test_r2']):.4f} ± {np.std(scores['test_r2']):.4f}")
print()
print(f"MAE (Train):  {np.mean(scores['train_mae']):.4f} ± {np.std(scores['train_mae']):.4f}")
print(f"MAE (Val):    {np.mean(scores['val_mae']):.4f} ± {np.std(scores['val_mae']):.4f}")
print(f"MAE (Test):   {np.mean(scores['test_mae']):.4f} ± {np.std(scores['test_mae']):.4f}")

# Средняя Permutation Importance по всем фолдам
print("\n" + "="*70)
print("СРЕДНЯЯ PERMUTATION IMPORTANCE ПО ВСЕМ ФОЛДАМ (ТОП-10)")
print("="*70)

# Усредняем по всем фолдам
avg_perm_importance = all_permutation_importance[0].copy()
for i in range(1, len(all_permutation_importance)):
    avg_perm_importance['importance_mean'] += all_permutation_importance[i]['importance_mean'].values
    avg_perm_importance['importance_std'] += all_permutation_importance[i]['importance_std'].values

avg_perm_importance['importance_mean'] /= len(all_permutation_importance)
avg_perm_importance['importance_std'] /= len(all_permutation_importance)
avg_perm_importance = avg_perm_importance.sort_values('importance_mean', ascending=False)

print(avg_perm_importance.head(10).to_string(index=False))

# График топ-10 признаков
plt.figure(figsize=(12, 6))
top_10 = avg_perm_importance.head(10)
plt.barh(range(len(top_10)), top_10['importance_mean'].values,
         xerr=top_10['importance_std'].values, capsize=5)
plt.yticks(range(len(top_10)), top_10['feature'].values)
plt.xlabel('Permutation Importance')
plt.title('Топ-10 важных признаков (Permutation Importance)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# График лучших итераций по фолдам
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(scores['best_iteration']) + 1), scores['best_iteration'])
plt.xlabel('Fold')
plt.ylabel('Best Iteration')
plt.title('Лучшая итерация для каждого fold')
plt.xticks(range(1, len(scores['best_iteration']) + 1))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import GroupKFold
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.inspection import permutation_importance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Подготовка данных
df_r = df[:1000].copy()
X = df_r.drop(['Plotnost__avg', 'ts_group','ProchnostNaSzhatie__avg','alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg','ptr_fluidity_low'], axis=1)
# X = df_r[['System1:RZN.L3.EX.ZE.TZZE.Data.ZEActValTempZone8_2__avg',
#        'System1:RZN.L3.EX.ZE.TZZE.Data.ZEActValTempZone8_3__avg',
#        'System1:RZN.L3.EX.ZE.TZZE.Data.ZEActValTempZone8_4__avg',
#         'Nomenklatura']]
# X = df_r[['alcohol','System1:RZN.L3.PL.CAB1.Data.CAB1_SPEED__avg','Nomenklatura']]
y = df_r['Plotnost__avg']

# Группы для кросс-валидации
groups = df_r['pres'].values

# Преобразуем 'Nomenklatura' в категорический тип
X['Nomenklatura'] = X['Nomenklatura'].astype('category')

# Получаем индексы категориальных столбцов
categorical_features = [X.columns.get_loc('Nomenklatura')]

print(f"Признаки: {X.shape[1]}")
# print(f"Категориальные признаки: {categorical_features}")
print(f"Уникальные группы: {len(np.unique(groups))}")

# GroupKFold для кросс-валидации по группам
gkf = GroupKFold(n_splits=5)

scores = {
    'train_rmse': [], 'val_rmse': [], 'test_rmse': [],
    'train_r2': [], 'val_r2': [], 'test_r2': [],
    'train_mae': [], 'val_mae': [], 'test_mae': [],
    'best_iteration': []
}

# Для хранения Permutation Importance
all_permutation_importance = []

# Кросс-валидация
for fold, (train_val_idx, test_idx) in enumerate(gkf.split(X, y, groups)):
    print(f"\n{'='*70}")
    print(f"--- Fold {fold + 1} ---")
    print('='*70)

    # Разделяем на train_val и test
    X_train_val, X_test = X.iloc[train_val_idx], X.iloc[test_idx]
    y_train_val, y_test = y.iloc[train_val_idx], y.iloc[test_idx]
    groups_train_val = groups[train_val_idx]

    # Внутренний split для train и val
    gkf_inner = GroupKFold(n_splits=4)
    for train_idx, val_idx in gkf_inner.split(X_train_val, y_train_val, groups_train_val):
        X_train, X_val = X_train_val.iloc[train_idx], X_train_val.iloc[val_idx]
        y_train, y_val = y_train_val.iloc[train_idx], y_train_val.iloc[val_idx]
        break  # Берем только первый split

    # Создаем Pool для CatBoost
    train_pool = Pool(X_train, y_train, cat_features=categorical_features)
    val_pool = Pool(X_val, y_val, cat_features=categorical_features)
    test_pool = Pool(X_test, y_test, cat_features=categorical_features)


    # Обучаем модель с early stopping
    model = CatBoostRegressor(
        iterations=1000,  # Максимальное количество итераций
        learning_rate=0.1,
        depth=6,
        loss_function='RMSE',
        eval_metric='RMSE',
        early_stopping_rounds=50,  # Останавливаем если нет улучшений за 50 итераций
        random_state=42,
        verbose=False
    )

    model.fit(train_pool, eval_set=val_pool)

    # Получаем количество использованных итераций
    best_iteration = model.get_best_iteration()
    scores['best_iteration'].append(best_iteration)

    # Предсказания
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)
    y_test_pred = model.predict(X_test)

    # Метрики
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    train_r2 = r2_score(y_train, y_train_pred)
    val_r2 = r2_score(y_val, y_val_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    train_mae = mean_absolute_error(y_train, y_train_pred)
    val_mae = mean_absolute_error(y_val, y_val_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)

    scores['train_rmse'].append(train_rmse)
    scores['val_rmse'].append(val_rmse)
    scores['test_rmse'].append(test_rmse)
    scores['train_r2'].append(train_r2)
    scores['val_r2'].append(val_r2)
    scores['test_r2'].append(test_r2)
    scores['train_mae'].append(train_mae)
    scores['val_mae'].append(val_mae)
    scores['test_mae'].append(test_mae)

    print(f"Best iteration: {best_iteration}")
    print(f"Train RMSE: {train_rmse:.4f} | Val RMSE: {val_rmse:.4f} | Test RMSE: {test_rmse:.4f}")
    print(f"Train R²:   {train_r2:.4f} | Val R²:   {val_r2:.4f} | Test R²:   {test_r2:.4f}")
    print(f"Train MAE:  {train_mae:.4f} | Val MAE:  {val_mae:.4f} | Test MAE:  {test_mae:.4f}")

    # Permutation Importance на тестовом наборе
    print(f"\nРассчитываю Permutation Importance...")
    perm_importance = permutation_importance(
        model, X_test, y_test,
        n_repeats=10,
        random_state=42,
        n_jobs=-1
    )

    perm_importance_df = pd.DataFrame({
        'feature': X.columns,
        'importance_mean': perm_importance.importances_mean,
        'importance_std': perm_importance.importances_std
    }).sort_values('importance_mean', ascending=False)

    all_permutation_importance.append(perm_importance_df)

    print(f"\nТоп-10 важных признаков (Permutation Importance) для Fold {fold + 1}:")
    print(perm_importance_df.head(10).to_string(index=False))

# Итоговые результаты
print("\n" + "="*70)
print("ИТОГОВЫЕ РЕЗУЛЬТАТЫ КРОСС-ВАЛИДАЦИИ")
print("="*70)
print(f"Лучшая итерация: {np.mean(scores['best_iteration']):.1f} ± {np.std(scores['best_iteration']):.1f}")
print()
print(f"RMSE (Train): {np.mean(scores['train_rmse']):.4f} ± {np.std(scores['train_rmse']):.4f}")
print(f"RMSE (Val):   {np.mean(scores['val_rmse']):.4f} ± {np.std(scores['val_rmse']):.4f}")
print(f"RMSE (Test):  {np.mean(scores['test_rmse']):.4f} ± {np.std(scores['test_rmse']):.4f}")
print()
print(f"R² (Train):   {np.mean(scores['train_r2']):.4f} ± {np.std(scores['train_r2']):.4f}")
print(f"R² (Val):     {np.mean(scores['val_r2']):.4f} ± {np.std(scores['val_r2']):.4f}")
print(f"R² (Test):    {np.mean(scores['test_r2']):.4f} ± {np.std(scores['test_r2']):.4f}")
print()
print(f"MAE (Train):  {np.mean(scores['train_mae']):.4f} ± {np.std(scores['train_mae']):.4f}")
print(f"MAE (Val):    {np.mean(scores['val_mae']):.4f} ± {np.std(scores['val_mae']):.4f}")
print(f"MAE (Test):   {np.mean(scores['test_mae']):.4f} ± {np.std(scores['test_mae']):.4f}")

# Средняя Permutation Importance по всем фолдам
print("\n" + "="*70)
print("СРЕДНЯЯ PERMUTATION IMPORTANCE ПО ВСЕМ ФОЛДАМ (ТОП-10)")
print("="*70)

# Усредняем по всем фолдам
avg_perm_importance = all_permutation_importance[0].copy()
for i in range(1, len(all_permutation_importance)):
    avg_perm_importance['importance_mean'] += all_permutation_importance[i]['importance_mean'].values
    avg_perm_importance['importance_std'] += all_permutation_importance[i]['importance_std'].values

avg_perm_importance['importance_mean'] /= len(all_permutation_importance)
avg_perm_importance['importance_std'] /= len(all_permutation_importance)
avg_perm_importance = avg_perm_importance.sort_values('importance_mean', ascending=False)

print(avg_perm_importance.head(10).to_string(index=False))

# График топ-10 признаков
plt.figure(figsize=(12, 6))
top_10 = avg_perm_importance.head(10)
plt.barh(range(len(top_10)), top_10['importance_mean'].values,
         xerr=top_10['importance_std'].values, capsize=5)
plt.yticks(range(len(top_10)), top_10['feature'].values)
plt.xlabel('Permutation Importance')
plt.title('Топ-10 важных признаков (Permutation Importance)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# График лучших итераций по фолдам
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(scores['best_iteration']) + 1), scores['best_iteration'])
plt.xlabel('Fold')
plt.ylabel('Best Iteration')
plt.title('Лучшая итерация для каждого fold')
plt.xticks(range(1, len(scores['best_iteration']) + 1))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
