```python
import pandas as pd
import numpy as np

# Создание DataFrame
data = {
    'SKU': [
        'Сыр полутвёрдый Bonfesto Pizza Моцарелла Пленка 250 г 40,0%',
        'Сыр полутвёрдый Hochland Grunlander Пленка 400 г 50,0%',
        'Сыр полутвёрдый LiebenDorf Гауда Пленка 1000 г 45,0%',
        'Сыр полутвёрдый LiebenDorf Маасдам Пленка 1000 г 45,0%',
        'Сыр полутвёрдый LiebenDorf Тильзитер Пленка 1000 г 45,0%'
    ],
    '2023-12-01 00:00:00': [101.75, 1729.6, 42943.2159999999, 55588.469, 62122.559],
    '2024-01-01 00:00:00': [129, np.nan, 44166.604, 41133.619, np.nan],
    '2024-02-01 00:00:00': [80.5, np.nan, np.nan, np.nan, 44]
}

df = pd.DataFrame(data)

# Функция для вычисления суммы
def compute_sum(row):
    if pd.isna(row['2024-02-01 00:00:00']):
        return np.nan
    else:
        return row[['2024-01-01 00:00:00', '2024-02-01 00:00:00']].sum(skipna=True)

# Создание новой колонки 'summ'
df['summ'] = df.apply(compute_sum, axis=1)

# Вывод результата
result = df[['summ']]
print(result)
```
