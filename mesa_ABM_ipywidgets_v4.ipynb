```python
import pandas as pd
import numpy as np
from mesa import Agent, Model
from mesa.time import RandomActivation
from sklearn.preprocessing import MinMaxScaler
from scipy.optimize import minimize

# Нормализуем признаки для удобства вычислений
scaler = MinMaxScaler()
df[['PRICE PER UNIT_norm', 'Analytic1_encoded_norm']] = scaler.fit_transform(
    df[['PRICE PER UNIT', 'Analytic1_encoded']]
)

# Класс агента
class ConsumerAgent(Agent):
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.choice = None

    def softmax(self, x):
        """ Применение softmax-функции для вычисления вероятностей """
        e_x = np.exp(x - np.max(x))  # Для численной устойчивости
        return e_x / e_x.sum(axis=0)

    def step(self):
        # Агенты получают вероятности выбора для каждого продукта из модели
        probabilities = self.model.probabilities

        # Выбираем продукт на основе вероятностей
        best_product_idx = np.random.choice(self.model.df.index, p=probabilities)
        self.choice = best_product_idx

        # Регистрируем выбор в модели
        self.model.register_choice(self.choice)

# Класс модели
class MarketModel(Model):
    def __init__(self, df, num_agents):
        super().__init__()
        self.num_agents = num_agents
        self.schedule = RandomActivation(self)
        self.df = df.copy()
        self.choice_counts = {idx: 0 for idx in self.df.index}

        # Создаем агентов
        for i in range(self.num_agents):
            agent = ConsumerAgent(i, self)
            self.schedule.add(agent)

        # Вычисляем вероятности для выбора SKU на основе характеристик
        self.probabilities = self.calculate_probabilities()

    def calculate_probabilities(self):
        """ Вычисление вероятностей для выбора SKU """
        # Определение оценок для каждого SKU на основе характеристик
        scores = []
        for idx, product in self.df.iterrows():
            price_score = -product['PRICE PER UNIT_norm']
            analytic_score = product['Analytic1_encoded_norm']

            # Общая оценка товара
            score = price_score + analytic_score
            scores.append(score)

        # Применение softmax для преобразования оценок в вероятности
        probabilities = self.softmax(np.array(scores))
        return probabilities

    def softmax(self, x):
        """ Применение softmax-функции """
        e_x = np.exp(x - np.max(x))  # Для численной устойчивости
        return e_x / e_x.sum(axis=0)

    def step(self):
        self.schedule.step()

    def register_choice(self, choice_idx):
        self.choice_counts[choice_idx] += 1

# Функция потерь для оптимизации
def loss_function(probabilities, df, num_agents):
    # Применяем нормализацию, чтобы вероятности оставались в пределах [0, 1]
    probabilities = np.array(probabilities)
    probabilities = probabilities / probabilities.sum()

    # Задаем вероятности для модели
    model = MarketModel(df, num_agents)
    model.probabilities = probabilities

    # Запускаем шаг модели
    model.step()

    # Получаем выборы агентов
    choice_counts = model.choice_counts
    df['agent_sku'] = df.index.map(choice_counts)

    # Вычисляем 'abm_sales_vol' и 'abm_share_vol'
    df['abm_sales_vol'] = df['agent_sku'] * df['abm_Analytic1']
    total_abm_sales_vol = df['abm_sales_vol'].sum()
    if total_abm_sales_vol == 0:
        return np.inf  # Избегаем деления на ноль
    df['abm_share_vol'] = df['abm_sales_vol'] / total_abm_sales_vol

    # Вычисляем функцию потерь как сумму квадратов разностей
    loss = np.sum((df['share_vol'] - df['abm_share_vol']) ** 2)

    return loss

# Количество агентов
num_agents = 1000

# Начальное приближение вероятностей для выбора SKU
initial_probabilities = np.random.rand(len(df))
initial_probabilities = initial_probabilities / initial_probabilities.sum()  # Нормализация

# Ограничения для вероятностей (все вероятности в пределах [0, 1])
bounds = [(0, 1)] * len(df)

# Запускаем оптимизацию
result = minimize(
    loss_function,
    initial_probabilities,
    args=(df, num_agents),
    bounds=bounds,
    method='L-BFGS-B'
)

# Получаем оптимизированные вероятности
optimized_probabilities = result.x
optimized_probabilities = optimized_probabilities / optimized_probabilities.sum()  # Нормализация

# Выводим оптимизированные вероятности
print("Оптимизированные вероятности выбора SKU:", optimized_probabilities)
print("Ошибка функции потерь:", result.fun)

# Запускаем модель с оптимизированными вероятностями
model = MarketModel(df, num_agents)
model.probabilities = optimized_probabilities
model.step()

# Обновляем датафрейм с результатами
choice_counts = model.choice_counts
df['agent_sku'] = df.index.map(choice_counts)
df['abm_sales_vol'] = df['agent_sku'] * df['abm_Analytic1']
total_abm_sales_vol = df['abm_sales_vol'].sum()
df['abm_share_vol'] = df['abm_sales_vol'] / total_abm_sales_vol if total_abm_sales_vol != 0 else 0

# Выводим обновленный датафрейм
df[['Sku_name', 'share_vol', 'abm_share_vol', 'agent_sku', 'abm_sales_vol']]
```
