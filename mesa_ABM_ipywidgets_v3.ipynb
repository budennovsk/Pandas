```python
import pandas as pd
import numpy as np
from mesa import Agent, Model
from mesa.time import RandomActivation
from sklearn.preprocessing import MinMaxScaler
from scipy.optimize import minimize

# # Ваш исходный датафрейм
# data = {
#     'Month': [29, 119, 179, 239, 269],
#     'Year': [2023, 2023, 2023, 2023, 2023],
#     'PRICE PER UNIT': [469.424564, 449.267694, 816.116336, 720.726262, 580.349508],
#     'SALES VOL': [76620.4, 41569.2, 31459.0, 34252.0, 43662.4],
#     'SHARE VOL': [0.244997, 0.132919, 0.243361, 0.264967, 0.139612],
#     'Analytic4': ['Ванночки', 'Ванночки', 'Ванночки', 'Ванночки', 'Ванночки'],
#     'Analytic1': ['400 г', '400 г', '200 г', '200 г', '400 г'],
#     'Brand': ['LiebenDorf', 'Свежий ряд', 'Фетакса', 'Hochland', 'Viola'],
#     'Sku_name': [
#         'Сыр плавленый LiebenDorf Сливочный Ванночки 400 г',
#         'Сыр плавленый Свежий ряд Янтарь Ванночки 400 г',
#         'Сыр плавленый Фетакса Ванночки 200 г 45,0%',
#         'Сыр плавленый Hochland Сливочный Ванночки 200 г',
#         'Сыр плавленый Viola Сливочный Ванночки 400 г 50%'
#     ],
#     'Analytic1_encoded': [6, 6, 4, 4, 6],
#     'share_vol': [0.173770, 0.094276, 0.071347, 0.077681, 0.099023],
#     'abm_Analytic1': [400, 400, 200, 200, 400]
# }

# df = pd.DataFrame(data)

df = df_12
df['abm_Analytic1'] = df['Analytic1'].str.replace(' г', '').astype(int)

# Нормализуем признаки для удобства вычислений
scaler = MinMaxScaler()
df[['PRICE PER UNIT_norm', 'Analytic1_encoded_norm']] = scaler.fit_transform(
    df[['PRICE PER UNIT', 'Analytic1_encoded']]
)

# Класс агента с коэффициентами, передаваемыми из модели
class ConsumerAgent(Agent):
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.choice = None
        # Коэффициенты будут установлены во время каждого шага
        self.coeff_price = None
        self.coeff_weight = None

    def step(self):
        # Агенты получают коэффициенты из модели
        self.coeff_price = self.model.coeff_price[self.unique_id]
        self.coeff_weight = self.model.coeff_weight[self.unique_id]

        # Агенты оценивают товары и выбирают лучший
        scores = []
        for idx, product in self.model.df.iterrows():
            # Рассчитываем оценку товара с использованием коэффициентов агента
            # score = (
            #     self.coeff_price * (product['PRICE PER UNIT_norm']) +
            #     self.coeff_weight * product['Analytic1_encoded_norm']
            # )

            price_score = -abs(self.coeff_price -product['PRICE PER UNIT_norm'])#-product['PRICE PER UNIT'] * self.price_sensitivity


            # Оценка аналитического параметра
            analytic_score = -abs(self.coeff_weight-product['Analytic1_encoded_norm'])

            # Общая оценка продукта
            score = price_score + analytic_score
            scores.append((idx, score))
        # Выбираем товар с максимальным баллом
        best_product_idx = max(scores, key=lambda x: x[1])[0]
        self.choice = best_product_idx
        # Регистрируем выбор в модели
        self.model.register_choice(self.choice)

# Класс модели с коэффициентами агентов
class MarketModel(Model):
    def __init__(self, df, num_agents, coeffs):
        super().__init__()
        self.num_agents = num_agents
        self.schedule = RandomActivation(self)
        self.df = df.copy()
        self.choice_counts = {idx: 0 for idx in self.df.index}

        # Разделяем коэффициенты на коэффициенты для цены и веса для каждого агента
        coeff_price = coeffs[:num_agents]
        coeff_weight = coeffs[num_agents:]
        self.coeff_price = coeff_price
        self.coeff_weight = coeff_weight

        # Создаем агентов
        for i in range(self.num_agents):
            agent = ConsumerAgent(i, self)
            self.schedule.add(agent)

    def step(self):
        self.schedule.step()

    def register_choice(self, choice_idx):
        self.choice_counts[choice_idx] += 1

# Функция потерь для оптимизации
def loss_function(coeffs, df, num_agents):
    # Инициализируем и запускаем модель с текущими коэффициентами
    model = MarketModel(df, num_agents, coeffs)
    model.step()
    # Получаем выборы агентов
    choice_counts = model.choice_counts
    df['agent_sku'] = df.index.map(choice_counts)
    # Вычисляем 'abm_sales_vol' и 'abm_share_vol'
    df['abm_sales_vol'] = df['agent_sku'] * df['abm_Analytic1']
    total_abm_sales_vol = df['abm_sales_vol'].sum()
    if total_abm_sales_vol == 0:
        return np.inf  # Избегаем деления на ноль
    df['abm_share_vol'] = df['abm_sales_vol'] / total_abm_sales_vol
    # print(df.to_string())
    # Вычисляем функцию потерь как сумму квадратов разностей
    loss = np.sum((df['share_vol'] - df['abm_share_vol']) ** 2)

    return -loss



num_agents = 10
# Ограничения для коэффициентов (от 0 до 1)
bounds = [(0, 1)] *num_agents*2  # 20 агентов * 2 коэффициента каждый

# Начальное приближение коэффициентов
initial_coeffs = np.random.rand(num_agents*2)

# Количество агентов


# Запускаем оптимизацию
result = minimize(
    loss_function,
    initial_coeffs,
    args=(df, num_agents),
    bounds=bounds,
    method='L-BFGS-B'
)

# Получаем оптимизированные коэффициенты
optimized_coeffs = result.x
optimized_coeff_price = optimized_coeffs[:num_agents]
optimized_coeff_weight = optimized_coeffs[num_agents:]

# Выводим оптимизированные коэффициенты
print("Оптимизированные coeff_price:", optimized_coeff_price)
print("Оптимизированные coeff_weight:", optimized_coeff_weight)
print()
print("Ошибка функции потерь:", -result.fun)

# Запускаем модель с оптимизированными коэффициентами
model = MarketModel(df, num_agents, optimized_coeffs)
model.step()

# Обновляем датафрейм с результатами
choice_counts = model.choice_counts
df['agent_sku'] = df.index.map(choice_counts)
df['abm_sales_vol'] = df['agent_sku'] * df['abm_Analytic1']
total_abm_sales_vol = df['abm_sales_vol'].sum()
df['abm_share_vol'] = df['abm_sales_vol'] / total_abm_sales_vol if total_abm_sales_vol != 0 else 0

# Выводим обновленный датафрейм
df[['Sku_name', 'share_vol', 'abm_share_vol', 'agent_sku', 'abm_sales_vol']]
```
